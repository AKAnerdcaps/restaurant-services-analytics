```python
import pandas as pd
import numpy as np
import html
import re
pd.set_option('display.max_columns', None)
```

```python
inspections_data = pd.read_csv('./data/raw/currentOpenInspections.csv')
```

```python
#1 Remove Duplicates
inspections_data = inspections_data.drop_duplicates()

#2 Capitalize Seating Interest Column
inspections_data['SeatingChoice'] = inspections_data['SeatingChoice'].apply(lambda x: x.capitalize())

inspections_data
```

```python
#3 Remove HTML encoding from columns (RestaurantName, LegalBusinessName, BusinessAddress)
inspections_data['RestaurantName'] = inspections_data['RestaurantName'].apply(lambda x: html.unescape(str(x)))
inspections_data['LegalBusinessName'] = inspections_data['LegalBusinessName'].apply(lambda x: html.unescape(str(x)))
inspections_data['BusinessAddress'] = inspections_data['BusinessAddress'].apply(lambda x: html.unescape(str(x)))
```

```python
#4 Capitalize RestaurantName, LegalBusinessName and BusinessAddress columns and standardize business suffixes and address abbreviations
def standardize_columns(df, column):
    df[column] = df[column].str.title()

    # Fix inconsistencies with Business Suffixes
    mapping = {
        r'\bLlc\b': 'LLC',
        r'\bInc(?!\.)\b': 'Inc.',
        r'\bIncorporated(?!\.)\b': 'Inc.',
        r'\bCorp(?!\.)\b': 'Corp.'
    }
    df[column] = df[column].replace(mapping, regex=True)

    # Fix issues with capitalization of contractions('s & 't become 'S and 'T after using title)
    mapping = {
        r"'S\b": "'s",
        r"'T\b": "'t"
    }
    df[column] = df[column].replace(mapping, regex=True)

    #Fix location abbreviations
    mapping = {
        r'\bSt$|\bSt\.$': "Street",
        r'\bAv$|\bAv\.$|\bAve$|\bAve\.$': "Avenue",
        r'\bLn$|\bLn\.$': "Lane",
        r'\bBlvd$|\bBlvd\.$|\bBoul$|\bBoul\.$': "Boulevard"
    }
    df[column] = df[column].replace(mapping, regex=True)

    return df

inspections_data = standardize_columns(inspections_data, 'RestaurantName')
inspections_data = standardize_columns(inspections_data, 'LegalBusinessName')
inspections_data = standardize_columns(inspections_data, 'BusinessAddress')

inspections_data
```

```python
#5 Convert suffixes Th,Rd,St,Nd to th,rd,st,nd respectively
def convert_suffixes(df, column):
    df[column] = df[column].str.replace(r'(\d+)(St|Nd|Rd|Th)\b', lambda match: match.group(1) + match.group(2).lower(), regex = True)
    return df

inspections_data = convert_suffixes(inspections_data, 'RestaurantName')
inspections_data = convert_suffixes(inspections_data, 'LegalBusinessName')
inspections_data = convert_suffixes(inspections_data, 'BusinessAddress')

inspections_data
```

```python
#6 Extracting Building Number from BusinessAddress column and rename BusinessAddress to Street
def extract_building_number(df):
    if df['BusinessAddress'].split()[0].isdigit():
        bld_num = df['BusinessAddress'].split()[0]
        df['Building Number'] = bld_num
        df['BusinessAddress'] = df['BusinessAddress'].replace(bld_num, '', 1).strip()
    else:
        df['Building Number'] = np.nan
    return df

inspections_data = inspections_data.apply(extract_building_number, axis=1)
inspections_data.rename(columns={'BusinessAddress': 'Street'}, inplace=True)

inspections_data
```

```python
#7 Convert standalone numbers to ordinals in Street i.e. 1->1st, 2->2nd, etc.
def convert_to_ordinal(match):
    number = int(match.group(0))

    if 10 <= number % 100 <= 20:
        suffix = 'th'
    else:
        suffix = {1: 'st', 2: 'nd', 3: 'rd'}.get(number % 10, 'th')
    return str(number) + suffix

# Make Ordinal Numbers from Words, i.e. First->1st, Second->2nd, etc.
def words_to_ordinal(word):
    word_to_number = {'First': '1st', 'Second': '2nd', 'Third': '3rd', 'Fourth': '4th', 'Fifth': '5th', 'Sixth': '6th' , 'Seventh': '7th', 'Eight': '8th', \
                      'Ninth': '9th', 'Tenth': '10th', 'Eleventh': '11th', 'Twelfth': '12th', 'Thirteenth': '13th', 'Fourteenth': '14th', 'Fifteenth': '15th'}
    return word_to_number.get(word, word)

inspections_data['Street'] = inspections_data['Street'].str.replace(r'\b(\d+)\b', convert_to_ordinal, regex=True)
inspections_data['Street'] = inspections_data['Street'].apply(lambda x: ' '.join([words_to_ordinal(word) for word in x.split()]))
```

```python
#8 Recreate Business Address column
inspections_data['Business Address'] = inspections_data['Building Number'] + ' ' + inspections_data['Street']+ ', ' + inspections_data['Borough']+ ', NY'
inspections_data
```
