```python
import pandas as pd
import numpy as np
import html
pd.set_option('display.max_columns', None)
```

```python
dohmh_data = pd.read_csv('./data/raw/DOHMH_New_York_City_Restaurant_Inspection_Results_20231202.csv')
```

```python
#1 Remove Duplicates
dohmh_data = dohmh_data.drop_duplicates()

#2 Fill missing values in zipcode with 0 and convert column to integer
dohmh_data['ZIPCODE'] = dohmh_data['ZIPCODE'].fillna(0).astype(int)

dohmh_data
```

```python
#3 Remove rows having NaN for DBA and STREET and 0 for BORO
dohmh_data = dohmh_data.dropna(subset=['DBA']).copy()
dohmh_data = dohmh_data.dropna(subset=['STREET']).copy()
dohmh_data = dohmh_data[dohmh_data['BORO']!='0']

dohmh_data
```

```python
#4 Remove HTML encoding from columns (DBA, STREET, ACTION, VIOLATION DESCRIPTION)
columns_to_unescape = ['DBA', 'STREET', 'STREET', 'VIOLATION DESCRIPTION']

def unescape_html(element):
    return html.unescape(element) if pd.notna(element) else element

for column in columns_to_unescape:
    dohmh_data[column] = dohmh_data[column].apply(unescape_html)

dohmh_data
```

```python
#5 Capitalize DBA and STREET columns and standardize business suffixes and address abbreviations
def standardize_columns(df, column):
    df[column] = df[column].fillna('').str.title()

    # Fix inconsistencies with Business Suffixes
    mapping = {
        r'\bLlc\b': 'LLC',
        r'\bInc(?!\.)\b': 'Inc.',
        r'\bIncorporated(?!\.)\b': 'Inc.',
        r'\bCorp(?!\.)\b': 'Corp.'
    }
    df[column] = df[column].replace(mapping, regex=True)

    # Fix issues with capitalization of contractions('s & 't become 'S and 'T after using title)
    mapping = {
        r"'S\b": "'s",
        r"'T\b": "'t"
    }
    df[column] = df[column].replace(mapping, regex=True)

    #Fix location abbreviations
    mapping = {
        r'\bSt$|\bSt\.$': "Street",
        r'\bAv$|\bAv\.$|\bAve$|\bAve\.$': "Avenue",
        r'\bLn$|\bLn\.$': "Lane",
        r'\bBlvd$|\bBlvd\.$|\bBoul$|\bBoul\.$': "Boulevard"
    }
    df[column] = df[column].replace(mapping, regex=True)

    return df

dohmh_data = standardize_columns(dohmh_data, 'DBA')
dohmh_data = standardize_columns(dohmh_data, 'STREET')

dohmh_data
```

```python
#6 Convert suffixes Th,Rd,St,Nd to th,rd,st,nd respectively
def convert_suffixes(df, column):
    df[column] = df[column].str.replace(r'(\d+)(St|Nd|Rd|Th)\b', lambda match: match.group(1) + match.group(2).lower(), regex = True)
    return df

dohmh_data = convert_suffixes(dohmh_data, 'STREET')
dohmh_data = convert_suffixes(dohmh_data, 'DBA')

dohmh_data
```

```python
#7 Convert standalone numbers to ordinals in Street i.e. 1->1st, 2->2nd, etc.
def convert_to_ordinal(match):
    number = int(match.group(0))

    if 10 <= number % 100 <= 20:
        suffix = 'th'
    else:
        suffix = {1: 'st', 2: 'nd', 3: 'rd'}.get(number % 10, 'th')
    return str(number) + suffix

# Make Ordinal Numbers from Words, i.e. First->1st, Second->2nd, etc.
def words_to_ordinal(word):
    word_to_number = {'First': '1st', 'Second': '2nd', 'Third': '3rd', 'Fourth': '4th', 'Fifth': '5th', 'Sixth': '6th' , 'Seventh': '7th', 'Eight': '8th', \
                      'Ninth': '9th', 'Tenth': '10th', 'Eleventh': '11th', 'Twelfth': '12th', 'Thirteenth': '13th', 'Fourteenth': '14th', 'Fifteenth': '15th'}
    return word_to_number.get(word, word)

dohmh_data['STREET'] = dohmh_data['STREET'].str.replace(r'\b(\d+)\b', convert_to_ordinal, regex=True)
dohmh_data['STREET'] = dohmh_data['STREET'].apply(lambda x: ' '.join([words_to_ordinal(word) for word in x.split()]))

dohmh_data
```

```python
dohmh_data['Business Address'] = dohmh_data['BUILDING'] + " "+ dohmh_data['STREET'] + ", " +dohmh_data['BORO'] + ", NY"
dohmh_data
```
